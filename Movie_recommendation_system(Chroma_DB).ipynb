```python
import numpy as np
import pandas as pd
import ast
```

```python
movies= pd.read_csv('/content/tmdb_5000_movies.csv')
credits= pd.read_csv('/content/tmdb_5000_credits.csv')
```

```python
movies = movies.merge(credits,on='title')
```

```python
#genres id keywords title overview cast crew
#selecting important columns for creating tags
movies = movies[['movie_id','title','overview','genres','keywords','cast','crew']]
```

```python
movies.dropna(inplace=True)
```

```python
def convert(obj):
    L=[]
    for i in ast.literal_eval(obj):
        L.append(i['name'])
    return L
```

```python
movies['genres'] = movies['genres'].apply(convert)
```

```python
movies['keywords'] = movies['keywords'].apply(convert)
```

```python
def convert3(obj):
    L=[]
    counter=0
    for i in ast.literal_eval(obj):
        if counter!=3:
            L.append(i['name'])
            counter+=1
        else:
            break
    return L
```

```python
movies['cast'] = movies['cast'].apply(convert3)
```

```python
def director_find(obj):
    L=[]
    for i in ast.literal_eval(obj):
        if i['job'] == 'Director':
            L.append(i['name'])
            break
    return L
```

```python
movies['crew'] = movies['crew'].apply(director_find)
```

```python
movies['overview'][0]
```

```python
movies['overview'] = movies['overview'].apply(lambda x:x.split())
```

```python
movies['genres'] = movies['genres'].apply(lambda x:[i.replace(" ","")for i in x])
movies['keywords'] = movies['keywords'].apply(lambda x:[i.replace(" ","")for i in x])
movies['cast'] = movies['cast'].apply(lambda x:[i.replace(" ","")for i in x])
movies['crew'] = movies['crew'].apply(lambda x:[i.replace(" ","")for i in x])
```

```python
movies['tags'] = movies['crew']+movies['cast']+movies['overview']+movies['genres']+movies['keywords']
```

```python
movies.head()
```

```python
new_df = movies[['movie_id','title','tags']]
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:" ".join(x))
```

```python
new_df['tags'] = new_df['tags'].apply(lambda x:x.lower())
```

```python
import chromadb
client = chromadb.PersistentClient(path="/content/")
```

```python
movies = []
metadatas = []
ids = []

# Iterate through each row in the DataFrame
for index, row in new_df.iterrows():
    # Extract relevant data from the DataFrame
    movie_id = row['movie_id']
    title = row['title']
    tags = row['tags']

    # Combine the title and tags as the movie content
    movie_content = f"{title} {tags}"

    # Append data to the respective lists
    movies.append(movie_content)
    metadatas.append({'movie_id': movie_id})
    ids.append(str(movie_id))  # Assuming 'movie_id' is unique and can be used as IDs
```

```python
# Create a ChromaDB collection for cosine similarity
movie_collection_cosine = client.create_collection(
    name="movie_collection_cosine",
    metadata={"hnsw:space": "cosine"}
)
```

```python
batch_size = 100
for i in range(0, len(movies), batch_size):
    movie_collection_cosine.add(
        documents=movies[i:i+batch_size],
        metadatas=metadatas[i:i+batch_size],
        ids=ids[i:i+batch_size]
    )
```

```python
# Define the movie name as input
movie_name = "The Dark Knight Rises"  # Replace with the desired movie name

# Retrieve the corresponding row from the DataFrame based on the movie name
movie_row = new_df[new_df['title'] == movie_name]

# Extract the tags for the selected movie
tags = movie_row['tags'].values[0]

# Set the query text to be the tags for the selected movie
query_text = tags

# Perform the query with the updated query text
results_cosine = movie_collection_cosine.query(
    query_texts=[query_text],
    n_results=11  # Change this to 5 for the top 5 results
)
```

```python
# Extract the recommended movie IDs from results_cosine
recommended_movie_ids = results_cosine['ids'][0]

# Create a list to store the recommended movie names
recommended_movie_names = []

# Map the movie IDs to movie names in new_df
for movie_id in recommended_movie_ids[1:]:
    movie_name = new_df[new_df['movie_id'] == int(movie_id)]['title'].values[0]
    recommended_movie_names.append(movie_name)

# Print the recommended movie names
recommended_movie_names
```

```python
import gradio as gr
import pandas as pd

# Assuming you have loaded your movie data into a DataFrame called 'new_df'

# Get a list of all movie names from the DataFrame
all_movie_names = new_df['title'].tolist()

def recommend_movies(movie_name):
    # Retrieve the corresponding row from the DataFrame based on the selected movie name
    movie_row = new_df[new_df['title'] == movie_name]

    # Extract the tags for the selected movie
    tags = movie_row['tags'].values[0]

    # Set the query text to be the tags for the selected movie
    query_text = tags

    # Perform the query with the updated query text
    results_cosine = movie_collection_cosine.query(
        query_texts=[query_text],
        n_results=10  # Change this to 5 for the top 5 results
    )

    # Extract the recommended movie IDs from results_cosine
    recommended_movie_ids = results_cosine['ids'][0]

    # Create a list to store the recommended movie names
    recommended_movie_names = []

    # Map the movie IDs to movie names in new_df
    for movie_id in recommended_movie_ids[1:]:
        movie_name = new_df[new_df['movie_id'] == int(movie_id)]['title'].values[0]
        recommended_movie_names.append(movie_name)

    return recommended_movie_names

# Create a dropdown input with all movie names
movie_name_input = gr.inputs.Dropdown(choices=all_movie_names)

iface = gr.Interface(fn=recommend_movies, inputs=movie_name_input, outputs="text")

iface.launch(share=True)
```
